```
from bs4 import BeautifulSoup
import urllib.request
import re
```

```
url = 'https://www.autocosmos.com.mx/clasificados/busqueda/usados/?aa=2015-2019'
```

```
page = urllib.request.urlopen(url) # conntect to website
```

```
soup = BeautifulSoup(page, 'html.parser')
print(soup)
```

```
mydivs = soup.findAll("article", {"class": "card listing-card m-clasificado"})
```

```
mydivs
```

```
brands =[]
models = []
model_dates = []
mileages = []
cities = []
prices = []
links = []

for i in mydivs:
    #the brand
    brand = i.find('span', itemprop="brand" ).text
    brands.append(brand)
    #the model
    model = i.find('div', itemprop="model").text
    models.append(model)
    #model dates
    dates = i.find('span',itemprop="modelDate").text
    model_dates.append(dates)
    #mileage
    kms = i.find('span',itemprop="mileageFromOdometer").text
    mileages.append(kms)
    #cities
    city = i.find('span',itemprop="addressLocality").text
    cities.append(city)
    #prices
    price = i.find('span',itemprop="price").text
    prices.append(price)
    link = i.find('a',itemprop="url").text
    links.append('https://www.autocosmos.com.mx'+link)
    
```

```
import pandas as pd
df = pd.DataFrame({'brand': brands,
'model': models,
'model_date': model_dates,
'mileage': mileages,
'city': cities,
'price': prices,
'url' : links                   
                   
})
df
```

```
pages = [str(i) for i in range(2,15)]
pages
```

```
from time import time
from IPython.core.display import clear_output
from warnings import warn
from time import sleep
from random import randint
from requests import get

brands =[]
models = []
model_dates = []
mileages = []
cities = []
prices = []
links = []

# Preparing the monitoring of the loop
start_time = time()
requests = 0


# For every page in the interval 1-4
for page in pages:

    # Make a get request
    response = urllib.request.urlopen('https://www.autocosmos.com.mx/clasificados/busqueda/usados/?aa=2015-2019&pageindex='+page)

    # Pause the loop
    sleep(randint(8,15))

    # Monitor the requests
    requests += 1
    elapsed_time = time() - start_time
    print('Request:{}; Frequency: {} requests/s'.format(requests, requests/elapsed_time))
    clear_output(wait = True)

    # Throw a warning for non-200 status codes
    
    # Break the loop if the number of requests is greater than expected
    if requests > 72:
        warn('Number of requests was greater than expected.')
        break

    # Parse the content of the request with BeautifulSoup
    soup = BeautifulSoup(response, 'html.parser')
    # Select all the 50 movie containers from a single page
    mydivs = soup.findAll("article", {"class": "card listing-card m-clasificado"})

    # For every movie of these 50
    for i in mydivs:
      brand = i.find('span', itemprop="brand" ).text
      brands.append(brand)
      #the model
      model = i.find('div', itemprop="model").text
      models.append(model)
      #model dates
      dates = i.find('span',itemprop="modelDate").text
      model_dates.append(dates)
      #mileage
      kms = i.find('span',itemprop="mileageFromOdometer").text
      mileages.append(kms)
      #cities
      city = i.find('span',itemprop="addressLocality").text
      cities.append(city)
      #price
      price = i.find('span',itemprop="price").text
      prices.append(price)
      link = i.find('a',itemprop="url").text
      links.append('https://www.autocosmos.com.mx'+link)
```

```
df2 = pd.DataFrame({'brand': brands,
'model': models,
'model_date': model_dates,
'mileage': mileages,
'city': cities,
'price': prices
                   
})
df2
```
